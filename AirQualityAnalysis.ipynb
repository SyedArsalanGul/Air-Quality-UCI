```python
subset_Dataset = Dataset
print (subset_Dataset)
```

```python
subset_Dataset["CO(GT)"] = subset_Dataset["CO(GT)"].replace(np.NaN, subset_Dataset["CO(GT)"].mean())
subset_Dataset["PT08.S1(CO)"] = subset_Dataset["PT08.S1(CO)"].replace(np.NaN, subset_Dataset["PT08.S1(CO)"].mean())
subset_Dataset["NMHC(GT)"] = subset_Dataset["NMHC(GT)"].replace(np.NaN, subset_Dataset["NMHC(GT)"].mean())
subset_Dataset["C6H6(GT)"] = subset_Dataset["C6H6(GT)"].replace(np.NaN, subset_Dataset["C6H6(GT)"].mean())
subset_Dataset["PT08.S2(NMHC)"] = subset_Dataset["PT08.S2(NMHC)"].replace(np.NaN, subset_Dataset["PT08.S2(NMHC)"].mean())
subset_Dataset["NOx(GT)"] = subset_Dataset["NOx(GT)"].replace(np.NaN, subset_Dataset["NOx(GT)"].mean())
subset_Dataset["PT08.S3(NOx)"] = subset_Dataset["PT08.S3(NOx)"].replace(np.NaN, subset_Dataset["PT08.S3(NOx)"].mean())
subset_Dataset["NO2(GT)"] = subset_Dataset["NO2(GT)"].replace(np.NaN, subset_Dataset["NO2(GT)"].mean())
subset_Dataset["PT08.S4(NO2)"] = subset_Dataset["PT08.S4(NO2)"].replace(np.NaN, subset_Dataset["PT08.S4(NO2)"].mean())
subset_Dataset["PT08.S5(O3)"] = subset_Dataset["PT08.S5(O3)"].replace(np.NaN, subset_Dataset["PT08.S5(O3)"].mean())
subset_Dataset["T"] = subset_Dataset["T"].replace(np.NaN, subset_Dataset["T"].mean())
subset_Dataset["RH"] = subset_Dataset["RH"].replace(np.NaN, subset_Dataset["RH"].mean())
subset_Dataset["AH"] = subset_Dataset["AH"].replace(np.NaN, subset_Dataset["AH"].mean())
```

```python
missing_values_count= subset_Dataset.isnull().sum()
print (missing_values_count)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.tree import DecisionTreeRegressor
model_1 = DecisionTreeRegressor()
model_1.fit(train_X, train_y)


y_pred1 = model_1.predict(val_X)


r2_score(val_y, y_pred1)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_2 = RandomForestRegressor(n_estimators = 10, random_state = 0)
model_2.fit(X, y)


#from sklearn.model_selection import train_test_split
#train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state = 0)
#model_2 = RandomForestRegressor()

model_2.fit(train_X, train_y)

y_pred2 = model_2.predict(val_X)


r2_score(val_y, y_pred2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 4)
poly_X = poly_reg.fit_transform(train_X)
model_3 = LinearRegression()
model_3.fit(poly_X, train_y)

y_pred3 = model_3.predict(poly_reg.transform(val_X))



r2_score(val_y, y_pred3)
```

```python
from sklearn.linear_model import LinearRegression
model_4 = LinearRegression()
model_4.fit(train_X, train_y)


y_pred4 = model_4.predict(val_X)


r2_score(val_y, y_pred4)
```

```python
from sklearn.svm import SVR
model_5 = SVR(kernel = 'rbf')
model_5.fit(train_X, train_y)


y_pred5 = model_5.predict(val_X)

r2_score(val_y, y_pred5)
```
